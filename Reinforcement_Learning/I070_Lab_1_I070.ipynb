```python
# Cell 1: Package Installation & Colab Setup
!pip install gymnasium matplotlib numpy pillow

import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

print(f"Gymnasium Version: {gym.__version__}")
```

```python
# Cell 2: CartPole-v1 - Standard vs. Customized Simulation
def run_cartpole_demo(custom_gravity=9.8, custom_masspole=0.1):
    # 1. Create Environment
    env = gym.make("CartPole-v1", render_mode="rgb_array")

    # 2. Customize parameters via unwrapped attribute
    env.unwrapped.gravity = custom_gravity
    env.unwrapped.masspole = custom_masspole
    env.unwrapped.total_mass = env.unwrapped.masspole + env.unwrapped.masscart
    env.unwrapped.polemass_length = env.unwrapped.masspole * env.unwrapped.length

    obs, info = env.reset(seed=42)
    print(f"=== CartPole-v1 (Gravity={custom_gravity}, MassPole={custom_masspole}) ===")
    print(f"Initial State Observation: {obs}")
    print(f"Observation Space: {env.observation_space}")
    print(f"Action Space: {env.action_space}\n")

    total_reward = 0
    steps = 0

    for _ in range(200):
        action = env.action_space.sample() # Random action selection
        obs, reward, terminated, truncated, info = env.step(action)
        total_reward += reward
        steps += 1

        if terminated or truncated:
            reason = "Terminated (Pole fell / Out of bounds)" if terminated else "Truncated (Max steps)"
            print(f"Episode Ended at Step {steps} | Reason: {reason}")
            break

    env.close()
    return steps, total_reward

# Run Default vs Custom
print("--- Standard Parameters ---")
run_cartpole_demo(custom_gravity=9.8, custom_masspole=0.1)

print("\n--- Customized Parameters (High Gravity & Heavy Pole) ---")
run_cartpole_demo(custom_gravity=25.0, custom_masspole=0.8)
```

```python
# Cell 3: MountainCar-v0 - Standard vs. Customized Simulation
def run_mountaincar_demo(custom_force=0.001):
    env = gym.make("MountainCar-v0", render_mode="rgb_array")

    # Customize motor push force
    env.unwrapped.force = custom_force

    obs, info = env.reset(seed=42)
    print(f"=== MountainCar-v0 (Force={custom_force}) ===")
    print(f"Initial State Observation [Position, Velocity]: {obs}")
    print(f"Action Space: {env.action_space}")

    total_reward = 0
    steps = 0

    for _ in range(200):
        # Rule-based policy: move right if velocity > 0, else push left to gain momentum
        action = 2 if obs[1] > 0 else 0
        obs, reward, terminated, truncated, info = env.step(action)
        total_reward += reward
        steps += 1

        if terminated or truncated:
            reason = "Terminated (Goal reached!)" if terminated else "Truncated (Time limit)"
            print(f"Episode Finished at Step {steps} | Reason: {reason}")
            break

    env.close()
    return steps, total_reward

# Run Default vs Custom
print("--- Standard Motor Force (0.001) ---")
run_mountaincar_demo(custom_force=0.001)

print("\n--- Customized High Motor Force (0.005) ---")
run_mountaincar_demo(custom_force=0.005)
```

```python
# Cell 4: Pendulum-v1 - Continuous State & Action Space
def run_pendulum_demo(custom_g=10.0, custom_l=1.0):
    env = gym.make("Pendulum-v1", render_mode="rgb_array")

    # Customize gravity and pendulum length
    env.unwrapped.g = custom_g
    env.unwrapped.l = custom_l

    obs, info = env.reset(seed=42)
    print(f"=== Pendulum-v1 (g={custom_g}, l={custom_l}) ===")
    print(f"Initial Observation [cos(theta), sin(theta), theta_dot]: {obs}")
    print(f"Action Space (Continuous Torque): {env.action_space}")

    total_reward = 0
    steps = 0

    for _ in range(200):
        action = env.action_space.sample() # Sample continuous torque array in [-2.0, 2.0]
        obs, reward, terminated, truncated, info = env.step(action)
        total_reward += reward
        steps += 1

        if terminated or truncated:
            break

    print(f"Completed {steps} steps | Cumulative Reward: {total_reward:.2f}")
    env.close()

print("--- Standard Pendulum Parameters ---")
run_pendulum_demo(custom_g=10.0, custom_l=1.0)

print("\n--- Custom Pendulum (Higher Gravity & Longer Rod) ---")
run_pendulum_demo(custom_g=19.6, custom_l=2.0)
```

```python
import gymnasium as gym
import numpy as np

def run_experiment(env_id, custom_params=None):
    env = gym.make(env_id)
    if custom_params:
        for key, val in custom_params.items():
            setattr(env.unwrapped, key, val)

    obs, info = env.reset(seed=42)
    total_reward = 0
    steps = 0

    while True:
        action = env.action_space.sample()
        obs, reward, terminated, truncated, info = env.step(action)
        total_reward += reward
        steps += 1
        if terminated or truncated:
            break

    env.close()
    return steps, total_reward

# 1. CartPole-v1
cp_default = run_experiment("CartPole-v1")
cp_custom = run_experiment("CartPole-v1", {"gravity": 25.0, "masspole": 0.5})

# 2. MountainCar-v0
mc_default = run_experiment("MountainCar-v0")
mc_custom = run_experiment("MountainCar-v0", {"force": 0.005})

# 3. Pendulum-v1
pen_default = run_experiment("Pendulum-v1")
pen_custom = run_experiment("Pendulum-v1", {"g": 19.6, "l": 2.0})

print("=== EXPERIMENTAL RESULTS ===")
print(f"CartPole-v1    | Default Steps: {cp_default[0]} | Custom (High Gravity) Steps: {cp_custom[0]}")
print(f"MountainCar-v0 | Default Steps: {mc_default[0]} | Custom (High Force) Steps: {mc_custom[0]}")
print(f"Pendulum-v1   | Default Reward: {pen_default[1]:.2f} | Custom (High Gravity) Reward: {pen_custom[1]:.2f}")
```

Helper Function to display

```python
import gymnasium as gym
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML

def display_simulation(frames, fps=30):
    """
    Renders a list of RGB frames as an HTML5 video player inside Google Colab.
    """
    fig, ax = plt.subplots(figsize=(6, 4))
    plt.axis('off')

    patch = ax.imshow(frames[0])
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

    def animate(i):
        patch.set_data(frames[i])
        return [patch]

    anim = animation.FuncAnimation(
        fig,
        animate,
        frames=len(frames),
        interval=1000/fps,
        blit=True
    )
    plt.close()
    return HTML(anim.to_html5_video())
```

Simulation 1: CartPole-v1

```python
# Create environment with rgb_array render mode
env = gym.make("CartPole-v1", render_mode="rgb_array")
obs, info = env.reset(seed=42)

frames = []
for _ in range(150):
    # Capture frame
    frames.append(env.render())

    # Take random action
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)

    if terminated or truncated:
        break

env.close()

# Display playable simulation video
print(f"Captured {len(frames)} frames. Rendering simulation...")
display_simulation(frames)
```

Simulation 2: MountainCar-v0

```python
env = gym.make("MountainCar-v0", render_mode="rgb_array")
obs, info = env.reset(seed=42)

frames = []
for _ in range(150):
    frames.append(env.render())

    # Heuristic action: push right if moving right, push left if moving left
    action = 2 if obs[1] > 0 else 0
    obs, reward, terminated, truncated, info = env.step(action)

    if terminated or truncated:
        break

env.close()

print(f"Captured {len(frames)} frames. Rendering simulation...")
display_simulation(frames)
```

Simulation 3: Pendulum-v1

Comparing Custom Parameters Visually

```python
import gymnasium as gym
import numpy as np

# Create the Pendulum environment
env = gym.make("Pendulum-v1", render_mode="rgb_array")
obs, info = env.reset(seed=42)

frames = []

for step in range(200):
    frames.append(env.render())

    # State representation: obs = [cos(theta), sin(theta), angular_velocity]
    angular_velocity = obs[2]

    # Energy Pumping Policy:
    # 1. If it's stationary at the bottom, kick it with max torque.
    # 2. If it's already moving, push hard in the direction of movement (sign of angular velocity).
    if abs(angular_velocity) < 0.1:
        action = np.array([2.0], dtype=np.float32)  # Initial kick
    else:
        # Apply full torque (+2.0 or -2.0) matching current spin direction
        action = np.array([np.sign(angular_velocity) * 2.0], dtype=np.float32)

    obs, reward, terminated, truncated, info = env.step(action)

env.close()

print(f"Captured {len(frames)} frames! Rendering pendulum spin simulation...")
display_simulation(frames)
```
